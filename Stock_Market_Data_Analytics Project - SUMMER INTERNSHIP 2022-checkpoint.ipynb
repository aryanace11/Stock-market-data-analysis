```python
# DAILY PERCENTAGE CHANGE
# r_t = (p_t/p_t-1) -1
# This defines r_t(return at time t) as equal to the price at time t divided by the price at time t-1 minus 1.
```

```python
# The greater the value of r_t either negative or positive the greater will be the risk and reward.
# This indicates a more volatile stock there is a lot of fluctuation
```

```python
# The lesser the value of r_t the lesser will be the risk and lesser will be the reward.
# This indicates a more stable stock there is a less fluctuation
```

```python
# tesla['returns'] = (tesla['Close']/tesla['Close'].shift(1)) -1
```

```python
tesla['returns'] = tesla['Close'].pct_change(1)
```

```python
tesla.head()
```

```python
ford['returns'] = ford['Close'].pct_change(1)
gm['returns'] = gm['Close'].pct_change(1)
```

```python
# CUMMULATIVE RETURN 
```

```python
# A CUMMULATIVE RETURN IS THE AGGREGATE AMOUNT AN INVESTMENT HAS GAINED OR LOST OVER TIME
# THIS IS INDEPENDENT OF THE PERIOD OF TIME INVOLVED
# WHILE DAILY RETURNS ARE USEFUL, IT DOESN'T GIVE INVESTORS AN IMMEDIATE INSIGHT INTO GAINS HE OR SHE MADE TILL DATE, ESPECIALLY IF STOCK IS VERY VOLATILE.
# CUMMULATIVE RETURN(CR) IS COMPUTED RELATIVE TO THE DAY INVESTMENT IS MADE.
# IF CR>1 YOU ARE MAKING PROFITS ELSE YOU ARE IN LOSS.
```

```python
tesla['Cummulative Return'] = (1+ tesla['returns']).cumprod()
```

```python
tesla
```

```python
ford['Cummulative Return'] = (1+ ford['returns']).cumprod()
gm['Cummulative Return'] = (1+ gm['returns']).cumprod()
```

```python
ford
```

```python
gm
```

```python
tesla['Cummulative Return'].plot(label='Tesla',figsize=(16,8))
ford['Cummulative Return'].plot(label='ford',figsize=(16,8))
gm['Cummulative Return'].plot(label='gm',figsize=(16,8))
plt.legend(loc=0)
```
